## 2. Resumir con enfoque en envío y entrega

**Línea nueva agregada:**
> "enfocándote en cualquier aspecto que mencione el envío y la entrega del producto"

**Contexto:** En un ecommerce real, cada departamento necesita
información diferente de las reseñas.

El departamento de envíos no le interesa saber si el producto
es bonito o caro. Solo necesita saber qué pasó con la entrega
¿llegó a tiempo? ¿hubo problemas?.

Con prompts específicos podemos extraer solo la información
relevante para cada área del negocio.

**⚠️ Observación:** Aunque el resumen se enfocó en el envío,
también incluyó información del precio que no es relevante
para ese departamento. Esto nos lleva a la diferencia entre
"resumir" y "extraer" que veremos más adelante.

````python
prompt = f"""
Tu tarea es generar un resumen corto de una reseña de producto \
de un sitio de ecommerce para dar feedback al departamento \
de envíos.

Resume la reseña a continuación, delimitada por triple \
comillas invertidas, en máximo 30 palabras, enfocándote \
en cualquier aspecto que mencione el envío y la entrega \
del producto.

Reseña: ```{resena_producto}```
"""

response = get_completion(prompt)
print(response)
````

## 3. Resumir con enfoque en precio y valor

**Línea nueva agregada:**
> "enfocándote en cualquier aspecto relevante al precio y valor percibido"

**Contexto:** El departamento de precios necesita saber si
los clientes sienten que el producto vale lo que pagaron.
No le interesa el envío ni el diseño — solo si el precio
es justo o no.

````python
prompt = f"""
Tu tarea es generar un resumen corto de una reseña de producto \
de un sitio de ecommerce para dar feedback al departamento \
de precios, responsable de determinar el precio del producto.

Resume la reseña a continuación, delimitada por triple \
comillas invertidas, en máximo 30 palabras, enfocándote \
en cualquier aspecto relevante al precio y valor percibido \
del producto.

Reseña: ```{resena_producto}```
"""

response = get_completion(prompt)
print(response)
````

## 4. Extraer vs Resumir

**Concepto clave:**
- **Resumir** → el modelo genera un resumen general, puede incluir información extra
- **Extraer** → el modelo saca ÚNICAMENTE la información específica pedida, nada más

Usaremos "extraer" para el departamento de envíos y veremos
la diferencia con el resumen anterior.

**⚠️ Observación:** "Extraer" debería ser más preciso que
"resumir" — solo la información específica pedida, sin nada extra.
En la práctica el modelo no siempre hace la distinción perfecta,
pero el concepto es importante para prompts más avanzados.

````python
prompt = f"""
Tu tarea es extraer información relevante de una reseña \
de producto de un sitio de ecommerce para dar feedback \
al departamento de envíos.

De la reseña a continuación, delimitada por triple \
comillas invertidas, extrae la información relevante \
al envío y la entrega. Limita a 30 palabras.

Reseña: ```{resena_producto}```
"""

response = get_completion(prompt)
print(response)
````

## 5. Resumir múltiples reseñas con un loop

**Contexto:** En un ecommerce real no tienes una sola reseña
sino cientos o miles. Procesar cada una manualmente es imposible.

Con un loop de Python podemos resumir todas las reseñas
automáticamente en segundos — esto es lo que hace la diferencia
entre usar el chat manualmente y construir una aplicación real.

````python
# Reseña 1 - el peluche panda que ya teníamos
resena_1 = resena_producto

# Reseña 2 - lámpara de dormitorio
resena_2 = """
Necesitaba una lámpara bonita para mi dormitorio, y esta \
tenía almacenamiento adicional y un precio razonable. \
Llegó rápido — en 2 días. El cordón se rompió durante \
el transporte y la empresa me envió uno nuevo con gusto. \
Llegó en pocos días también. Fue fácil de armar. Luego \
me faltó una pieza, contacté a soporte y me la enviaron \
muy rápido. Me parece una gran empresa que se preocupa \
por sus clientes y productos.
"""

# Reseña 3 - cepillo de dientes eléctrico
resena_3 = """
Mi higienista dental me recomendó un cepillo eléctrico, \
por eso lo compré. La duración de la batería parece \
bastante impresionante. Después de la carga inicial y \
dejarlo enchufado la primera semana, lo desenchufé y \
lo he usado dos veces al día durante 3 semanas con \
la misma carga. Pero el cabezal es muy pequeño — he \
visto cepillos de bebé más grandes. Me gustaría que \
fuera más grande con cerdas de diferente largo para \
llegar mejor entre los dientes. Si lo consigues por \
unos $50, es un buen trato. ¡Mis dientes se sienten \
increíblemente limpios!
"""

# Reseña 4 - licuadora
resena_4 = """
Tenían el sistema de 17 piezas en oferta por unos $49 \
en noviembre, casi a mitad de precio, pero en diciembre \
los precios subieron a entre $70-$89 por el mismo sistema. \
Se ve bien, pero la base donde encaja la cuchilla no \
se ve tan bien como en ediciones anteriores. Después \
de un año el motor empezó a hacer un ruido extraño. \
Llamé al servicio al cliente pero la garantía ya había \
vencido. La calidad general ha bajado en estos productos.
"""

resenas = [resena_1, resena_2, resena_3, resena_4]
````

````python
for i in range(len(resenas)):
    prompt = f"""
    Tu tarea es generar un resumen corto de una reseña \
    de producto de un sitio de ecommerce.

    Resume la reseña a continuación, delimitada por triple \
    comillas invertidas, en máximo 20 palabras.

    Reseña: ```{resenas[i]}```
    """
    response = get_completion(prompt)
    print(f"Reseña {i+1}:", response, "\n")
````
