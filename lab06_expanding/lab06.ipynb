````python
from helper import get_completion
````

## Generar email de respuesta personalizado

**Contexto:** Una empresa recibe cientos de reseñas diarias.
Responder cada una manualmente es imposible. Con un LLM
podemos generar respuestas personalizadas automáticamente
— usando detalles específicos de cada reseña y adaptando
el tono según el sentimiento del cliente.

````python
# Sentimiento extraído previamente (como vimos en el lab de inferir)
sentimiento = "negativo"

# Reseña de una licuadora
resena = f"""
Todavía tenían el sistema de 17 piezas en oferta por \
unos $49 en noviembre, casi a mitad de precio, pero por \
alguna razón en la segunda semana de diciembre los precios \
subieron a entre $70-$89 por el mismo sistema. El sistema \
de 11 piezas también subió unos $10 desde el precio de \
oferta de $29. Se ve bien, pero la base donde encaja la \
cuchilla no se ve tan bien como en ediciones anteriores. \
Después de un año el motor empezó a hacer un ruido extraño. \
Llamé al servicio al cliente pero la garantía ya había \
vencido, así que tuve que comprar otra. La calidad general \
ha bajado en estos productos.
"""
````

## 1. Email de respuesta con temperature=0
(respuesta predecible y consistente)

````python
prompt = f"""
Eres un asistente de IA de servicio al cliente.
Tu tarea es enviar una respuesta por email a un cliente valioso.
Dado el email del cliente delimitado por triple comillas, \
genera una respuesta para agradecerle su reseña.
Si el sentimiento es positivo o neutral, agradécele su reseña.
Si el sentimiento es negativo, discúlpate y sugiere que \
puede contactar al servicio al cliente.
Asegúrate de usar detalles específicos de la reseña.
Escribe en un tono conciso y profesional.
Firma el email como "Agente de IA de Servicio al Cliente".
Reseña del cliente: ```{resena}```
Sentimiento de la reseña: {sentimiento}
"""

response = get_completion(prompt, temperature=0)
print(response)
````

## 2. Mismo prompt con temperature=0.7
(respuesta más creativa y variable)

**Línea cambiada:**
> temperature=0 → temperature=0.7

**¿Qué esperar?** Cada vez que ejecutes obtendrás
una respuesta diferente — el modelo es más creativo
pero menos predecible.

````python
response = get_completion(prompt, temperature=0.7)
print(response)
````
