# Transformar Texto
En esta lección exploraremos cómo usar LLMs para tareas
de transformación de texto como traducción de idiomas,
corrección ortográfica y gramatical, cambio de tono
y conversión de formatos.

## Casos de uso reales:
- Traducir documentación técnica en inglés
- Convertir datos entre formatos (JSON → HTML, CSV → tabla)
- Corregir textos antes de publicarlos
- Adaptar el tono según la audiencia

````python
from helper import get_completion
````

## 1. Traducción

**Contexto:** Los LLMs fueron entrenados con texto de internet
en cientos de idiomas. Eso les da capacidad de traducción
sin necesidad de un traductor especializado.

````python
# Traduccion basica - ingles a español
prompt = f"""
Traduce el siguiente texto en inglés al español: \
```Hola, me gustaría ordenar una licuadora```
"""

response = get_completion(prompt)
print(response)
````

### Detectar idioma

**Contexto:** Útil cuando recibes mensajes de usuarios
en idiomas desconocidos y necesitas identificar cuál es
antes de procesarlo.

````python
# Detectar en que idioma esta escrito un texto
prompt = f"""
Dime en qué idioma está escrito esto:
```Combien coûte le lampadaire?```
"""

response = get_completion(prompt)
print(response)
````

### Traducir a múltiples idiomas a la vez

**Contexto:** Con un solo prompt puedes obtener la traducción
en varios idiomas simultáneamente — sin hacer múltiples llamadas.

````python
# Traducir a multiples idiomas en una sola llamada
prompt = f"""
Traduce el siguiente texto al francés, español
e inglés pirata: \
```Quiero ordenar una pelota de basketball```
"""

response = get_completion(prompt)
print(response)
````

### Formal e informal

**Contexto:** En español existe el "tú" (informal) y el "usted"
(formal). El modelo puede traducir en ambas formas según
la audiencia — útil para empresas que atienden distintos
tipos de clientes.

````python
# Traducir en forma formal e informal
prompt = f"""
Traduce el siguiente texto al español en forma
formal e informal:
'Would you like to order a pillow?'
"""

response = get_completion(prompt)
print(response)
````

## 2. Traductor Universal

**Contexto:** Una empresa multinacional recibe mensajes de
soporte técnico en muchos idiomas diferentes. El equipo
de soporte no habla todos esos idiomas.

Con un loop podemos detectar el idioma de cada mensaje
y traducirlo automáticamente al español e inglés.

````python
# Mensajes de usuarios en diferentes idiomas
mensajes_usuarios = [
    "La performance du système est plus lente que d'habitude.",  # Frances
    "Mi monitor tiene píxeles que no se iluminan.",              # Español
    "Il mio mouse non funziona",                                 # Italiano
    "Mój klawisz Ctrl jest zepsuty",                             # Polaco
    "我的屏幕在闪烁"                                               # Chino
]
````

````python
# Loop que detecta el idioma y traduce cada mensaje
for mensaje in mensajes_usuarios:
    prompt = f"Dime en qué idioma está esto: ```{mensaje}```"
    idioma = get_completion(prompt)
    print(f"Mensaje original ({idioma}): {mensaje}")

    prompt = f"""
    Traduce el siguiente texto al español e inglés: \
```{mensaje}```
    """
    response = get_completion(prompt)
    print(response, "\n")
````

## 3. Transformación de Tono

**Contexto:** El mismo mensaje puede expresarse de forma
muy diferente según la audiencia.

- **Informal/jerga** → para amigos o compañeros
- **Formal/profesional** → para clientes, jefes o profesores

El modelo puede convertir automáticamente entre estos tonos.

````python
# Convertir jerga informal a carta de negocios formal
prompt = f"""
Traduce lo siguiente de jerga informal a una carta de negocios formal:
'Ey bro, soy Juan, mira estas especificaciones de esta lámpara de pie.'
"""

response = get_completion(prompt)
print(response)
````

## 4. Conversión de Formatos

**Contexto:** En desarrollo es muy común necesitar convertir
datos entre formatos — JSON a HTML, CSV a tabla, XML a JSON.

Antes esto requería escribir código complejo con expresiones
regulares. Ahora un solo prompt lo hace automáticamente.

````python
# Convertir JSON a tabla HTML
datos_json = {"empleados_restaurante": [
    {"nombre": "Carlos", "email": "carlos@gmail.com"},
    {"nombre": "Ana", "email": "ana32@gmail.com"},
    {"nombre": "Luis", "email": "luis87@gmail.com"}
]}

prompt = f"""
Traduce el siguiente diccionario de Python de JSON a una \
tabla HTML con encabezados de columna y título: {datos_json}
"""

response = get_completion(prompt)
print(response)
````

````python
from IPython.display import display, HTML
display(HTML(response))
````

## 5. Corrección Ortográfica y Gramatical

**Contexto:** Antes de publicar cualquier texto — un post,
un email, una reseña — es importante revisarlo.

El modelo puede corregir errores ortográficos, gramaticales
y de puntuación automáticamente. Especialmente útil cuando
escribes en un idioma que no es tu lengua nativa.

````python
# Lista de textos con errores gramaticales y ortograficos
textos = [
    "La chica con los cachorros blancos y negros tiene una pelota.",  # correcto
    "Yolanda tiene su cuaderno.",  # correcto
    "Voy a ser un dia largo. Necesita el carro cambiar su aceite?",  # errores
    "Alla va mi libertad. Alla van a traer sus maletas.",  # errores
    "Tu vas a necesitar tu cuaderno.",  # errores
    "Ese medicamento afecta mi capacidad de dormir.",  # correcto
    "Esta frace es para checar ChatGPT para habilidad de ortografia"  # errores
]

for t in textos:
    prompt = f"""Revisa y corrige el siguiente texto y \
reescribe la versión corregida. Si no encuentras errores, \
simplemente di "No se encontraron errores". \
No uses puntuación alrededor del texto:
```{t}```"""
    response = get_completion(prompt)
    print(response)
````

### Corregir una reseña completa

**Contexto:** Antes de publicar una reseña en internet
es útil corregirla. El modelo no solo corrige errores
sino que puede mejorar el estilo y tono del texto.

````python
# Reseña con errores para corregir
texto = f"""
Compré esto para mi hija por su cumpleaños porque siempre \
me lo quitaba de mi cuarto. Si, los adultos también gustan \
de los pandas. Ella lo lleva a todos lados, y es super suave \
y bonito. Una de las orejas esta un poco mas abajo que la otra, \
y no creo que eso fue diseñado para ser asimetrico. Es un poco \
pequeño para lo que pague. Creo que podría haber otras opciones \
mas grandes por el mismo precio. Llego un día antes de lo esperado, \
así que pude jugar con el antes de dárselo a mi hija.
"""

prompt = f"Revisa y corrige esta reseña: ```{texto}```"
response = get_completion(prompt)
print(response)
````

````python
from redlines import Redlines
from IPython.display import display, Markdown

diff = Redlines(texto, response)
display(Markdown(diff.output_markdown))
````

### Corregir y mejorar el estilo

**Contexto:** Además de corregir errores, el modelo puede
mejorar el tono y estilo del texto para hacerlo más
convincente y profesional — útil para publicaciones,
emails importantes o reportes.

````python
prompt = f"""
Revisa y corrige esta reseña. Hazla más convincente.
Asegúrate de que siga un estilo formal y esté dirigida
a un lector avanzado. Presenta el resultado en formato markdown.
Texto: ```{texto}```
"""

response = get_completion(prompt)
display(Markdown(response))
````
